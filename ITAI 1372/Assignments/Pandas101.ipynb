```python
#Jeffery. Dirden
#Pandas 101
#ITAI-1372

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('PastHires.csv')

# Extract rows 5-10 and only the "Previous Employers" and "Hired" columns
subset_df = df[['Previous employers', 'Hired']][5:11]

# Display the subset DataFrame
print(subset_df)

# Create a histogram for the 'Previous employers' column
subset_df['Previous employers'].plot(kind='hist', bins=5, edgecolor='black')

# Add labels and title
plt.xlabel('Number of Previous Employers')
plt.ylabel('Frequency')
plt.title('Distribution of Previous Employers (Rows 5-10)')

# Display the plot
plt.show()
```
